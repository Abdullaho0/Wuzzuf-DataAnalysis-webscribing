# Wuzzuf DataAnaysis Jobs Web_scirping

```python
# Import packages
import requests
import re
from bs4 import BeautifulSoup
```

```python
# Define the url of the site
base_site = "https://wuzzuf.net/search/jobs/?a=hpb&q=data%20analyst&start="
```

```python
# Sending a request to the web page
response = requests.get(base_site + str(0) )
response.status_code
```

### It seems that the whole pages are loaded successfully

```python
# Get the HTML from the web page
html = response.content
```

```python
# Convert the HTML to baeutifulsoup object
soup = BeautifulSoup(html, "lxml") 
soup
```

```python
# Eporting the HTML to a file
with open('WUZZUF DATA ANALYSIS JOBS.html', 'wb') as file:
    file.write(soup.prettify('utf-8'))
```

```python
# Find all the a tages on the web page containing The companies names
a_tages = soup.find_all("a", {'class':"css-17s97q8"}) 
```

```python
# Getting the companies names that need a data analyst
company = [ a.string for a in a_tages] 
company
```

```python
# Find all the span tages that contains work places
spans = soup.find_all('span', {"class" : "css-5wys0k"})
spans
```

```python
# Extaracting the adresses
adress = [span.text for span in spans]
adress
```

```python
# Finding all the span tages that have the Job types
spans = soup.find_all("span", class_="css-1ve4b75 eoyjyou0")
spans
```

```python
# Extracting the Job types
job_type = [span.text for span in spans]
job_type
```

```python
# Finding all the div tages that contains since howmany days the job was released
divs = soup.find_all("div", class_ = ("css-do6t5g" or "4c4ojb") )
# Extracting the age 
job_age = [div.text for div in divs]
job_age
```

```python
# Finding all the a tages that contains the years of experinces required
spans = soup.find_all("span")
spans
```

```python
# Extracting the years of experince 
yrs_of_exp = []
for span in spans:
    if "Yrs of Exp" in span.text:
        yrs_of_exp.append(span.text)
        
yrs_of_exp
```

```python
# Getting the minimum experince years
min_yrs_of_exp = [int(re.findall(r'^\D*(\d+)', year)[0]) for year in yrs_of_exp]
min_yrs_of_exp
```

```python
# getting the whole pages of the web sit
for page in range(1, 14):
    response = requests.get(base_site + str(page) )
    html = response.content
    soup = BeautifulSoup(html, "lxml")
    a_tages = soup.find_all("a", {'class':"css-17s97q8"})
    # Extaracting the the company name
    for a in a_tages:
        company.append(a.string)
    # Find all the span tages that contains work places
    spans = soup.find_all('span', {"class" : "css-5wys0k"})
    # Extaracting the adresses
    for span in spans:
        adress.append(span.text)
    # Finding all the span tages that have the Job types
    spans = soup.find_all("span", class_="css-1ve4b75 eoyjyou0")
    # Extracting the Job types
    for span in spans:
        job_type.append(span.text)
    # Finding all the div tages that contains since howmany days the job was released
    divs = soup.find_all("div", class_ = ("css-do6t5g" or "4c4ojb") )
    # Extracting the age
    for div in divs:
        job_age.append(div.text)
    # Finding all the a tages that contains the years of experinces required
    spans = soup.find_all("span")
    # Extracting the years of experince 
    yrs_of_exp = []
    for span in spans:
        if "Yrs of Exp" in span.text:
            yrs_of_exp.append(span.text)
    # Getting the minimum experince years
    for year in yrs_of_exp:
        min_yrs_of_exp.append(int(re.findall(r'^\D*(\d+)', year)[0]) )
```

# Populating the DataFrame

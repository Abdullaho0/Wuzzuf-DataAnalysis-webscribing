# Investigate the WuzzuF DataAnalysis Jobs

```python
#Importing important packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
df = pd.read_csv('DataAnalysit_jobs_info.csv')
df.head()
```

```python
# Fristly we can unpackage the address to [City & Country]
country = [ df['address'][i].split(',')[-1] for i in range(0, len(df) ) ]
city = [ df['address'][i].split(',')[-2] for i in range(0, len(df) ) ]
print(len(country), len(city))
```

```python
# Add city & country to the data frame
df['city'] = city
df['country'] = country
df
```

```python
df.city.value_counts()
```

```python
# Removing the space if excists in the city name, to get the true unique values
city = [df['city'][i].replace(' ', '') for i in range(0,len(df))]
df['city'] = city
```

```python
city
```

```python
# check the unique values for the city columns
df.city.value_counts()
```

```python
# What is the most wandted experienced years
df.min_exp_yrs.value_counts()
```

```python
# Lets check the values for the job age
df.job_age.value_counts()
```

```python
# Lets make all the job_age only in days, tp make it easier in visualization
df['job_age'].replace({'1 month ago':'30 days ago', '2 months ago' : '60 days ago'}, inplace = True)
df.job_age.value_counts()
```

```python
# Lets make it as an integer of number of days instead
import re
job_age_in_days = [int(re.search(r'\d+', df.job_age[i]).group()) for i in range(0, len(df))]
df['job_age_in_days'] = job_age_in_days
df
```

```python
# Lets check the job_type column, to see if it needs any modifications
df.job_type.value_counts()
```
